```python
import cv2
from deepface import DeepFace
import matplotlib.pyplot as plt

# Carregar a imagem
img = cv2.imread("Img/medo.png")

# Redimensionar e converter para RGB
img = cv2.resize(img, (640, 480))
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']
# Detectar a emoo
result = DeepFace.analyze(img_rgb, actions=['emotion'])

# Obter a emoo dominante
dominant_emotion = result[0]['dominant_emotion']

# Desenhar um quadrado verde ao redor da imagem
# (x, y) so as coordenadas iniciais, e (x+w, y+h) as finais do retngulo
start_point = (50, 50)  # Coordenada superior esquerda do retngulo
end_point = (590, 430)  # Coordenada inferior direita do retngulo
color = (0, 255, 0)  # Verde em formato BGR
thickness = 2  # Espessura da linha

# Desenha o retngulo verde
cv2.rectangle(img, start_point, end_point, color, thickness)

# Adicionar a emoo dominante na imagem
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, dominant_emotion, (50, 40), font, 1, color, 2, cv2.LINE_AA)

# Exibir a imagem usando matplotlib
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Para remover os eixos
plt.show()
```
